```python
plt.imshow(img_rgb)
```

### OpenCVの特徴
- numpy形式で読み込まれる。
- BGRで読み込まれる。変換が必要。

# グレースケール変換

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

```python
plt.gray()
plt.imshow(img_gray) #matplotlibのグレスケが画像は宣言が必要
```

```python
type(img_gray), img_gray.shape
```

# エッジ検出フィルタをかけてみる

```python
kernel1 = np.array([
    [-1, 0, 1],
    [-1, 0, 1],
    [-1, 0, 1],
])
```

```python
img_conv1 = cv2.filter2D(img_gray, -1, kernel1)
```

```python
plt.gray()
plt.imshow(img_conv1)
```

```python
kernel2 = np.array([
    [1, 0, -1],
    [1, 0, -1],
    [1, 0, -1],
])
img_conv2 = cv2.filter2D(img_gray, -1, kernel2)
plt.gray()
plt.imshow(img_conv2)
```

```python
kernel3 = np.array([
    [1, 1, 1],
    [0, 0, 0],
    [-1, -1, -1],
])
img_conv3 = cv2.filter2D(img_gray, -1, kernel3)
plt.gray()
plt.imshow(img_conv3)
```

```python
kernel4 = np.array([
    [-1, -1, -1],
    [0, 0, 0],
    [1, 1, 1],
])
img_conv4 = cv2.filter2D(img_gray, -1, kernel4)
plt.gray()
plt.imshow(img_conv4)
```

```python
img_conv = (img_conv1 + img_conv2 + img_conv3 + img_conv4)/4
```

```python
plt.gray()
plt.imshow(img_conv)
```
