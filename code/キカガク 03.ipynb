# 入力変数と出力変数の切り分け

```python
df.head(3)
```

```python
#df.iloc[行、列]
X = df.iloc[:,:-1]
```

```python
y = df.iloc[:,-1]
```

```python
y
```

```python
# モデルの構築と検証
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression(normalize=True)
```

```python
model.fit(X, y)
```

```python
# 検証（決定係数の計算）
model.score(X,y)
```

# 予測値の計算

```python
x = X.iloc[0,:]
x
```

```python
#予測値の計算
y_pred = model.predict([x])
y_pred
```

# モデルの保存

```python
from sklearn.externals import joblib
```

```python
joblib.dump(model, 'model.pkl')
```

# モデルの読み込み

```python
model_new = joblib.load('model.pkl')
```

```python
model_new.predict([x])
```

# パラメータの確認

```python
# パラメータ w の値
# なぜ予測がうまくいったか　ー＞　モデルの変数wの重みが大きかったものがいい！！
model.coef_
```
